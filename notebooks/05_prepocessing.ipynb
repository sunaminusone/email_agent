## **Concating PST and EML**

```python
import pandas as pd
from pathlib import Path

df_pst_all = pd.read_parquet("../data/processed/pst_parsed/parsed_pst_all.parquet")

print(df_pst_all.shape)
df_pst_all.head()
```

```python
df_pst_all.info()
```

```python
df_eml_all = pd.read_parquet("../data/processed/eml_parsed/parsed_eml_all.parquet")

print(df_eml_all.shape)
df_eml_all.head()
```

```python
df_all = pd.concat([df_pst_all, df_eml_all], ignore_index=True)

print("Final shape:", df_all.shape)
df_all.head()
```

```python
df_all.to_parquet("../data/processed/df_all_emails.parquet", index=False)
print("saved.")
```
